```python
import networkx as nx # networkx library
```

```python
G=nx.Graph() # creating a graph
```

```python
G.add_edge(1,2) # adding an edge to a graph
```

```python
nx.draw_networkx(G) # printing graph
```

```python
G.add_edges_from([(1,2),(2,3),(3,1)]) # adding all the edges at a time
```

```python
nx.draw_networkx(G) # printing graph
```

```python
G.edges() # edges in a graph 
```
