```python
import networkx as nx # networkx library
```

```python
G=nx.Graph() # creating a graph
```

```python
G.add_node(1) # adding a single node in a graph
```

```python
G.add_nodes_from([2,3,4]) # adding multiple nodes at a time in a graph
```

```python
nx.draw_networkx(G) # printing the graph
```

```python
G.add_node('spam') # As node is a hashable object(A file, string, float or anything can be added as a node). A string can also be added as a node
```

```python
nx.draw_networkx(G)
```

```python
G.add_nodes_from('spam') # adding nodes as seperatley as s,p,a,m
```

```python
nx.draw_networkx(G) # printing the graph
```

```python
h=nx.path_graph(10) # adding nodes from 0 to 9 and adding edges to the nodes
```

```python
nx.draw_networkx(h) # printing the path graph
```

```python
G.add_nodes_from(h) # adding the nodes from another graph to this graph
```

```python
nx.draw_networkx(G) # printing graph
```

```python
h.add_nodes_from(G) # adding a graph to the other graph but the edges of graph G are not preserved
```

```python
nx.draw_networkx(h) # printing graph
```

```python
G.add_edge('s','p') # adding edge 
```
