# Creating a karate club network and experimenting

```python
import networkx as nx # networkx library
```

```python
G=nx.karate_club_graph() # creating a karate club network
```

```python
nx.draw_networkx(G) # printing the graph
```

```python
nx.average_shortest_path_length(G) # finding average shortest pathlength in the graph
```

```python
nx.eccentricity(G) # finding eccentricity
```

```python
nx.radius(G) # finding radius in a graph
```

```python
nx.diameter(G) # finding diameter in a graph
```

```python
nx.periphery(G) # finding periphery in a graph
```

```python
nx.center(G) # finding centre in a graph
```
